## ** **3. PHASE 2: DATA PREPROCESSING

## 3.1 Introduction

Data preprocessing is a crucial step in the data analysis process.
In this phase, the raw telecom customer dataset is cleaned and transformed into a structured format suitable for analysis.

This includes:

-Handling missing values
-Fixing data type issues
-Removing duplicate records
-Preparing features for analysis

## 3.2 Handling Missing Values
Objective:
Identify and treat missing values to ensure data completeness and accuracy.

```python
missing_values = df.isnull().sum()
print("Missing Values Summary:\n", missing_values)
```

# -Handling Missing Values
# -Numerical Columns
# -Converts TotalCharges to numeric
#-Replaces missing values with median to avoid skewness

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

 ## 3.3 Data Type Conversion
 Objective:
 Ensure all columns have correct and appropriate data types.

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

## 3.4 Removing Duplicate Records
Objective:
Eliminate redundant entries to improve data quality.

Identifies duplicate rows
Removes them to avoid biased analysis

```python
duplicates = df.duplicated().sum()
print("Duplicate Rows:", duplicates)

df.drop_duplicates(inplace=True)
```

## 3.5 Feature Engineering
Objective:

Prepare and transform features for better analysis.

```python
df['Tenure_Years'] = df['tenure'] / 12
```

## Data Consistency Checks
Objective:

-Ensure uniformity in categorical values.
-Removes inconsistencies in text data.
-Standardizes categorical values.

```python
df['gender'] = df['gender'].str.lower()
df['Contract'] = df['Contract'].str.strip()
```

## 3.7 Final Data Validation

Purpose:
-Ensure no missing values remain.
-Confirm correct data types.
-Validate cleaned dataset.

```python
df.info()
df.isnull().sum()
```

## 3.8 Saving Cleaned Dataset

```python
cleaned_df = df.copy()
cleaned_df.to_csv("Telco_Customer_Churn_Cleaned.csv", index=False)

print("Cleaned dataset saved successfully!")
```

# 4.2 Churn Distribution Analysis
## Objective: Understand how many customers have churned vs retained.

```python
df['Churn'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Churn', data=df)
plt.title("Customer Churn Distribution")
plt.show()
```

## Insights:
-Shows proportion of churned vs non-churned customers
-Helps identify class imbalance

# 4.3 Demographic Analysis
## Gender vs Churn

```python
sns.countplot(x='gender', hue='Churn', data=df)
plt.title("Churn by Gender")
plt.show()
```

## Insights:
-Compare churn behavior across genders
-Identify if gender influences churn

## Senior Citizen vs Churn

```python
sns.countplot(x='SeniorCitizen', hue='Churn', data=df)
plt.title("Churn by Senior Citizen")
plt.show()
```

## Insights:
-Senior citizens may show different churn patterns

# 4.4 Service Usage Analysis
## Internet Service vs Churn

```python
sns.countplot(x='InternetService', hue='Churn', data=df)
plt.title("Churn by Internet Service")
plt.show()
```

## Contract Type vs Churn

```python
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Churn by Contract Type")
plt.xticks(rotation=30)
plt.show()
```

## Insights:
-Customers with month-to-month contracts usually churn more
-Long-term contracts improve retention

# 4.5 Billing Analysis

-Monthly Charges vs Churn

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

## Total Charges vs Churn

```python
sns.boxplot(x='Churn', y='TotalCharges', data=df)
plt.title("Total Charges vs Churn")
plt.show()
```

## Insights:
-Higher monthly charges may lead to higher churn
-Spending patterns differ between churned and retained customers

# 4.6 Tenure Analysis
Tenure Distribution

## Tenure vs Churn

## Insights:

# 4.7 Multivariate Analysis

## Objective:
Understand interactions between multiple variables.

```python
## Churn by Contract & Payment Method
pd.crosstab(df['Contract'], df['PaymentMethod'])
```

```python
## Pivot Table Analysis

pivot_table = df.pivot_table(
    values='MonthlyCharges',
    index='Contract',
    columns='Churn',
    aggfunc='mean'
)

print(pivot_table)
```

## Insight:

# 4.8 GroupBy Analysis
